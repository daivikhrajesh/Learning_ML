```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/iris.csv')
```

```python
def distance(point1, point2):

  dimentionality = len(point1)
  dist = 0

  for i in range(dimentionality):
    dist +=  (point1[i] - point2[i])**2

  dist = dist ** (0.5)
  return dist



def dist_array(point, X, y):
  arr = []

  for i, p in enumerate(X):
    arr.append(distance(point, p))

  df_temp = pd.DataFrame(X)
  df_temp['y'] = y
  df_temp['dist'] = arr

  return df_temp.sort_values('dist')



def classification(df_temp, k):
  class_votes = {}
  df_temp2 = df_temp.head(k)
  data = df_temp2.values
  number_of_classes = len(np.unique(y))

  for class_ in range(number_of_classes):
    class_votes[class_] = 0

  for row in data:
    class_ = int(row[4])
    class_votes[class_] +=1

  return max(class_votes, key=class_votes.get)
```

```python
X = df[['sepal.length',	'sepal.width',	'petal.length',	'petal.width']].values
y = np.concatenate([np.zeros(50), np.ones(50), np.ones(50)*2]).astype('int')
```

```python
def kNN_classification(X, y, k, point):
  iris_species = ['Setosa', 'Versicolor', 'Virginica']

  dist = dist_array(point, X, y)
  class_ = classification(dist, k)

  return iris_species[class_]
```
