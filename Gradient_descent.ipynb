```python
import numpy as np

def grad(x,y):
    m_c, b_c = 0,0
    iterations = 10000
    n= len(x)
    learning_rate = 0.084
    
    for i in range(iterations):
        y_pred = m_c*x + b_c
        cost = (1/n) * sum ([val**2 for val in (y - y_pred)])
        m_dev = -(2/n) * sum(x * (y - y_pred))
        b_dev = -(2/n) * sum(y - y_pred)
        m_c = m_c - learning_rate * m_dev 
        b_c = b_c - learning_rate * b_dev
        print(f"m {m_c}, b {b_c},cost {cost}, iteration {i}")

x = np.array([1,2,3,4,5])
y = np.array([5,6,7,8,9])

grad(x,y)
```
