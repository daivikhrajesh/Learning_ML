```python
import cv2
```

```python
image = cv2.imread('goku.jpg')
```

```python
h, w = image.shape[:2]
```

```python
print("Height = {}, Width = {}".format(h, w))
```

```python
# Extracting RGB values.Here we have randomly chosen a pixel by passing in 100, 100 for height and width.

(B, G, R) = image[150, 100]

# Displaying the pixel values
print("R = {}, G = {}, B = {}".format(R, G, B))

# We can also pass the channel to extract the value for a specific channel
B = image[100, 100, 0]
print("B = {}".format(B))
```

```python
roi = image[100 : 500, 200 : 700]
cv2.imshow("ROI", roi)
cv2.waitKey(0)
```

```python
# resize() function takes 2 parameters,
# the image and the dimensions
resize = cv2.resize(image, (500, 500))
cv2.imshow("Resized Image", resize)
cv2.waitKey(0)
```

```python
# Calculating the ratio
ratio = 800 / w

# Creating a tuple containing width and height
dim = (800, int(h * ratio))

# Resizing the image
resize_aspect = cv2.resize(image, dim)
cv2.imshow("Resized Image", resize_aspect)
cv2.waitKey(0)
```

```python
# We are copying the original image,
# as it is an in-place operation.
output = image.copy()

# Using the rectangle() function to create a rectangle.
rectangle = cv2.rectangle(output, (1500, 900), (600, 400), (255, 0, 0), 2)

cv2.imshow('Image with Rectangle', rectangle)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
img=cv2.imread("goku.jpg")

plt.imshow(img)
```

```python
# cv2 uses BGR color format and matplotlib uses RGB color format.
```

```python
img = image.copy()
# Converting BGR color to RGB color format
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
 
#Displaying image using plt.imshow() method
plt.imshow(RGB_img)
```

```python
path = r'goku.jpg'
```

```python
# Reading an image in default mode
image = cv2.imread(path)# Window name in which image is displayed

window_name = 'Goku'

# Using cv2.imshow() method
# Displaying the image
cv2.imshow(window_name, img)

# waits for user to press any key
cv2.waitKey(0)

# closing all open windows
cv2.destroyAllWindows()
```

```python
# Reading an image in grayscale mode
image = cv2.imread(path, 0)

window_name = 'Goku'

# Using cv2.imshow() method
# Displaying the image
cv2.imshow(window_name, img)

# waits for user to press any key
cv2.waitKey(0)

# closing all open windows
cv2.destroyAllWindows()
```
