```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression as LR
import math
from sklearn.preprocessing import StandardScaler
```

```python
def sk_learn():
    df = pd.read_csv("test_scores.csv")
    r = LR()
    r.fit(df[['math']],df.cs)
    m_sk,b_sk = r.coef_, r.intercept_
    
    #m_sk, b_sk = model.coef_[0], model.intercept_
    return m_sk, b_sk
```

```python
def grad_des(x,y):
    
    m_cur,b_cur = 0, 0
    cost_cur = float('inf')
    iterations = 5000
    learning_rate = 0.0001
    n = len(x)
    
    for i in range(iterations):
        y_pred = m_cur * x + b_cur
        cost = (1/n) * sum([val**2 for val in (y-y_pred)])
        m_der = -(2/n) * sum(x * (y - y_pred))
        b_der = -(2/n) * sum(y - y_pred)
        m_cur = m_cur - learning_rate * m_der
        b_cur = b_cur - learning_rate * b_der
        
        if math.isclose(cost, cost_cur, rel_tol=1e-20):
            break
        cost_cur = cost
        print(f"m {m_cur}, b {b_cur} ,cost {cost}, iteration {i}")
        
    return m_cur, b_cur

#cost_cur = float('inf')  # Initialize cost_cur to a large value
    
```

```python
if __name__ == "__main__":
    df = pd.read_csv("test_scores.csv")
    x = np.array(df.math)
    y = np.array(df.cs)
    
    m,b = grad_des(x,y)
    print("\n")

    print(f"Using gradient descent function: Coef {m} Intercept {b}")
    
    m_sklearn,b_sklearn = sk_learn()
    print("\n")

    print(f"Using sklearn: Coef {m_sklearn} Intercept {b_sklearn}")
```
